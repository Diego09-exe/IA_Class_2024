# **Decision tree**

In this notebook a decision tree is trained with the Iris dataset. More info:
* https://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html
* https://archive.ics.uci.edu/dataset/53/iris



Step 1: Import libraries.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn import tree
```

Step 2: Load data

```python
breastc = load_breast_cancer()
X = breastc.data
y = breastc.target
```

```python
breastc.feature_names
```

Step 3: Preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Step 4: Select and train the model.

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
# Exactitud del modelo en el entrenamiento.
train_accuracy = model.score(X_train, y_train)
print(train_accuracy)
```

Step 5: Test the model.

```python
y_pred = model.predict(X_test)
```

```python
y_test
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100)
print("Classification Report:\n", classification_report(y_test, y_pred, target_names=breastc.target_names))
```

Decision tree visualization.

```python
plt.figure(figsize=(20, 15))
tree.plot_tree(model, filled=True, feature_names=breastc.feature_names, class_names=breastc.target_names)
plt.title("Decision Tree for Breast Cancer Wisconsin (Diagnostic)")
plt.savefig("decision_tree.png", dpi=10)
plt.show()
```
